```python
from transformers import AutoTokenizer, AutoModelForQuestionAnswering
import torch

# Cargar el tokenizador y el modelo
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-cased-distilled-squad")
model = AutoModelForQuestionAnswering.from_pretrained("distilbert-base-cased-distilled-squad")
```

```python
context = """
El Amazonas es la selva tropical más grande del mundo y abarca una vasta área de América del Sur. Es famosa por su increíble biodiversidad, albergando una gran variedad de especies de plantas y animales. El río Amazonas, el río más caudaloso del mundo, fluye a través de ella.
"""

question = "Cuál es la selva tropical más grande del mundo?"
```

```python
# Tokenizar el contexto y la pregunta
inputs = tokenizer(question, context, return_tensors="pt")

# Obtener las predicciones del modelo
with torch.no_grad():
    outputs = model(**inputs)

answer_start_scores = outputs.start_logits
answer_end_scores = outputs.end_logits


answer_start = torch.argmax(answer_start_scores)
answer_end = torch.argmax(answer_end_scores) + 1


answer = tokenizer.decode(inputs["input_ids"][0][answer_start:answer_end])

print(f"Pregunta: {question}")
print(f"Respuesta: {answer}")
```

```python
question_2 = "¿Qué río fluye a través de la selva amazónica?"
inputs_2 = tokenizer(question_2, context, return_tensors="pt")
with torch.no_grad():
    outputs_2 = model(**inputs_2)
answer_2 = tokenizer.decode(inputs_2["input_ids"][0][torch.argmax(outputs_2.start_logits):torch.argmax(outputs_2.end_logits)+1])
print(f"Pregunta: {question_2}")
print(f"Respuesta: {answer_2}")
```

```python
from transformers import AutoTokenizer, AutoModelForQuestionAnswering
import torch


print("Cargando el modelo de Preguntas y Respuestas. Esto puede tomar un momento...")
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-cased-distilled-squad")
model = AutoModelForQuestionAnswering.from_pretrained("distilbert-base-cased-distilled-squad")
print("Modelo cargado exitosamente. ¡Listo para tus preguntas!")
print("-" * 50)


while True:
    print("\n--- NUEVA PREGUNTA ---")
    context = input("Por favor, introduce el CONTEXTO (o escribe 'salir' para terminar): \n")
    if context.lower() == 'salir':
        print("¡Adiós!")
        break

    question = input("Ahora, introduce tu PREGUNTA basada en el contexto: \n")


    inputs = tokenizer(question, context, return_tensors="pt")


    with torch.no_grad():
        outputs = model(**inputs)


    answer_start_scores = outputs.start_logits
    answer_end_scores = outputs.end_logits


    answer_start = torch.argmax(answer_start_scores)
    answer_end = torch.argmax(answer_end_scores) + 1



    if answer_start >= answer_end:
        answer = "No pude encontrar una respuesta clara en el contexto proporcionado."
    else:
        answer = tokenizer.decode(inputs["input_ids"][0][answer_start:answer_end])

    print("\n--- RESULTADO ---")
    print(f"Pregunta: {question}")
    print(f"Respuesta: {answer}")
    print("-" * 50)
```
